```python
import matplotlib.pyplot as plt

import numpy as np

# Data
years = [2013, 2015, 2016, 2016, 2016, 2017, 2017, 2018, 2018, 2021, 2022, 2023]
competitions = [
    'WorldU18 Championships', 'Asian Championships', 'SouthAsian Games', 
    'AsianU20 Championships', 'WorldU20 Championships', 'Asian Championships', 
    'World Championships', 'Commonwealth Games', 'Asian Games', 
    'Olympic Games', 'World Championships', 'World Championships'
]
results = [66.75, 70.5, 82.23, 77.6, 86.48, 85.23, 82.26, 86.47, 88.06, 87.58, 88.13, 88.17]
positions = [19, 9, 1, 2, 1, 1, 15, 1, 1, 1, 2, 1]  # Add positions here
venue = ['Ukraine', 'China', 'India', 'Vietnam', 'Poland', 'India', 'UK', 'Australia', 'Indonesia', 'Japan', 'USA', 'Hungry']

# Create the line chart
plt.figure(figsize=(12, 8))
ax = plt.gca() 
ax.set_facecolor('white')
plt.gcf().set_facecolor('white')

for spine in ax.spines.values():
    spine.set_visible(False)

# Offsets for annotations
offsets = [(0, 35), (0, -20), (-35, 5), (0, -25), (-20, 30), (0, 35), (0, -20), (40, -20), (0, 30), (0, -25), (0, 35), (0, -25)]

# Plot each point and annotate with the competition name
for i in range(len(years)):
    color = 'lightgrey'  # Default color
    if positions[i] == 1:
        color = 'gold'
    elif positions[i] == 2:
        color = 'lightblue'
    elif positions[i] == 3:
        color = '#cd7f32'  # Bronze
    
    plt.plot(years, results, linestyle='--', color='lightgrey', linewidth=0.8, zorder =1)
    plt.scatter(years[i], results[i], color=color, s=150, zorder=2)
    
    # Annotate the competition name
    competition_text = competitions[i].replace(' ', '\n')
    plt.annotate(competition_text, (years[i], results[i]), textcoords="offset points", xytext=offsets[i], ha='center', va='center', fontname='Futura', fontsize=11)

# Annotate the distance in light grey color
    distance_text = f"{venue[i]} - {results[i]}m"
    plt.annotate(distance_text, (years[i], results[i]), textcoords="offset points", xytext=(offsets[i][0], offsets[i][1] - 20), ha='center', va='center', fontname='Futura', color='grey', fontsize=9)

    # for y in np.arange(int(min(results)) - 1, int(max(results)) + 5, 5):
# Add your handle name on the bottom right
plt.text(max(years)+1, min(results) - 5, '@VisualizeTrends', ha='right', va='bottom', fontname='Futura', fontsize=10, color='grey')

# Add the source as Wikipedia on the bottom left
plt.text(min(years)-1, min(results) - 4, 'Source: Olympics.com', ha='left', va='bottom', fontname='Futura', fontsize=10, color='grey')
plt.text(min(years)-1, min(results) - 5, 'Note: Graph does not include circuit titles', ha='left', va='bottom', fontname='Futura', fontsize=10, color='grey')

# Add labels and title
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='lightgrey', zorder=0)
plt.grid(axis='x', linestyle='--', linewidth=0.01, color='white', zorder=0)
plt.xlim(2012, 2024)
plt.xticks(range(2013, 2024), fontname='Futura')
plt.yticks(np.arange(int(min(results)) - 1, int(max(results)) + 2, 3), fontname='Futura')
plt.ylabel('Distance (m)', fontname='Futura')
plt.title('Neeraj Chopra: A Decade of Dominance in Javelin', fontname='Futura', fontweight='bold', pad=80, fontsize=30)

plt.legend(handles=[
    plt.Circle((0, 0), color='gold', label='Gold', fill=True),
    plt.Circle((0, 0), color='lightblue', label='Silver', fill=True),
    plt.Circle((0, 0), color='lightgrey', label='Other', fill=True)
], loc='upper center', bbox_to_anchor=(0.5, 1.18), fontsize=10, frameon=False, ncol=3)

plt.tight_layout()
plt.savefig('Neeraj.png', bbox_inches='tight', dpi=500, facecolor='white')


```
